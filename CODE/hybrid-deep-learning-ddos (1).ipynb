```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import joblib

# ==================== 1. DATA LOADING & PREPROCESSING ====================
data_path = '/kaggle/input/edgeiiotset-cyber-security-dataset-of-iot-iiot/Edge-IIoTset dataset/Selected dataset for ML and DL/ML-EdgeIIoT-dataset.csv'
df = pd.read_csv(data_path, low_memory=False)

print(f"Dataset shape: {df.shape}")
print(f"Attack label distribution:\n{df['Attack_label'].value_counts()}")

# Lowercase columns
df.columns = df.columns.str.lower()

# Drop problematic columns (as per original preprocessing)
drop_columns = [
    "frame.time", "ip.src_host", "ip.dst_host", "arp.src.proto_ipv4",
    "arp.dst.proto_ipv4", "http.file_data", "http.request.full_uri",
    "icmp.transmit_timestamp", "http.request.uri.query", "tcp.options",
    "tcp.payload", "tcp.srcport", "tcp.dstport", "udp.port", "mqtt.msg"
]
existing_drop_cols = [col for col in drop_columns if col in df.columns]
df = df.drop(existing_drop_cols, axis=1)
print(f"Dropped {len(existing_drop_cols)} columns")

# Encode categorical columns
object_cols = df.select_dtypes(include=['object']).columns
for col in object_cols:
    if col != 'attack_type':  # Keep attack_type for multi-class
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))

print(f"Final dataset shape: {df.shape}")

# ==================== 2. BINARY CLASSIFICATION (SIMPLE & EFFECTIVE) ====================
print("\n" + "="*60)
print("BINARY CLASSIFICATION")
print("="*60)

# Prepare data
X_bin = df.drop(['attack_label', 'attack_type'], axis=1)
y_bin = df['attack_label'].values

# Split
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(
    X_bin, y_bin, test_size=0.2, stratify=y_bin, random_state=42
)

# Scale
scaler_bin = StandardScaler()
X_train_bin_scaled = scaler_bin.fit_transform(X_train_bin)
X_test_bin_scaled = scaler_bin.transform(X_test_bin)

# Simple DNN for binary classification
def create_binary_model(input_dim):
    inputs = Input(shape=(input_dim,))
    x = Dense(128, activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(32, activation='relu')(x)
    outputs = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    )
    return model

binary_model = create_binary_model(X_train_bin_scaled.shape[1])
binary_model.summary()

# Train with early stopping
early_stop = EarlyStopping(patience=10, restore_best_weights=True)
history_bin = binary_model.fit(
    X_train_bin_scaled, y_train_bin,
    epochs=50,
    batch_size=256,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate
y_pred_bin = (binary_model.predict(X_test_bin_scaled) > 0.5).astype(int)
print(f"\nBinary Test Accuracy: {accuracy_score(y_test_bin, y_pred_bin):.4f}")
print(f"Binary F1-Score: {f1_score(y_test_bin, y_pred_bin):.4f}")
print("\nBinary Classification Report:")
print(classification_report(y_test_bin, y_pred_bin, target_names=['Normal', 'Attack']))

# ==================== 3. MULTI-CLASS CLASSIFICATION (IMPROVED) ====================
print("\n" + "="*60)
print("MULTI-CLASS CLASSIFICATION")
print("="*60)

# Create merged attack types column
# Strategy: Merge rare attacks and similar attack types
def merge_attack_types(attack_type):
    # Merge rare attacks
    if attack_type in ['MITM', 'Fingerprinting']:
        return 'Other_Rare'
    # Merge DDoS attacks
    elif attack_type in ['DDoS_HTTP', 'DDoS_TCP', 'DDoS_UDP', 'DDoS_ICMP']:
        return 'DDoS'
    # Merge web attacks
    elif attack_type in ['SQL_injection', 'XSS', 'Uploading']:
        return 'Web_Attack'
    # Merge scanning attacks
    elif attack_type in ['Port_Scanning', 'Vulnerability_scanner']:
        return 'Scanning'
    # Keep others as is
    else:
        return attack_type

df['attack_type_merged'] = df['attack_type'].apply(merge_attack_types)

# Encode the merged attack types
le_attack = LabelEncoder()
df['attack_type_encoded'] = le_attack.fit_transform(df['attack_type_merged'])
num_classes = len(le_attack.classes_)

print(f"Number of classes: {num_classes}")
print("Classes:", le_attack.classes_)

# Check distribution
attack_counts = df['attack_type_merged'].value_counts()
print("\nClass distribution after merging:")
for attack_type, count in attack_counts.items():
    print(f"{attack_type}: {count} ({count/len(df)*100:.2f}%)")

# Prepare data
X_multi = df.drop(['attack_label', 'attack_type', 'attack_type_merged', 'attack_type_encoded'], axis=1)
y_multi = df['attack_type_encoded'].values

# Split with stratification
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X_multi, y_multi, test_size=0.2, stratify=y_multi, random_state=42
)

# Scale
scaler_multi = StandardScaler()
X_train_multi_scaled = scaler_multi.fit_transform(X_train_multi)
X_test_multi_scaled = scaler_multi.transform(X_test_multi)

# Compute class weights for imbalanced data
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_multi),
    y=y_train_multi
)
class_weight_dict = dict(enumerate(class_weights))
print(f"\nClass weights: {class_weight_dict}")

# Enhanced model for multi-class
def create_multi_class_model(input_dim, num_classes):
    inputs = Input(shape=(input_dim,))
    
    # Branch 1: Main features
    x1 = Dense(256, activation='relu')(inputs)
    x1 = BatchNormalization()(x1)
    x1 = Dropout(0.4)(x1)
    x1 = Dense(128, activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = Dropout(0.3)(x1)
    
    # Branch 2: Statistical features
    x2 = Dense(64, activation='relu')(inputs)
    x2 = BatchNormalization()(x2)
    x2 = Dropout(0.2)(x2)
    
    # Combine branches
    combined = concatenate([x1, x2])
    combined = Dense(64, activation='relu')(combined)
    combined = BatchNormalization()(combined)
    combined = Dropout(0.3)(combined)
    
    outputs = Dense(num_classes, activation='softmax')(combined)
    
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

multi_class_model = create_multi_class_model(X_train_multi_scaled.shape[1], num_classes)
multi_class_model.summary()

# Enhanced training with callbacks
callbacks = [
    EarlyStopping(patience=15, restore_best_weights=True, monitor='val_loss'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
]

# Train with class weights
history_multi = multi_class_model.fit(
    X_train_multi_scaled, y_train_multi,
    epochs=100,
    batch_size=512,
    validation_split=0.2,
    callbacks=callbacks,
    class_weight=class_weight_dict,
    verbose=1
)

# Evaluate
y_pred_multi_probs = multi_class_model.predict(X_test_multi_scaled)
y_pred_multi = np.argmax(y_pred_multi_probs, axis=1)

print(f"\nMulti-Class Test Accuracy: {accuracy_score(y_test_multi, y_pred_multi):.4f}")
print(f"Multi-Class Macro F1-Score: {f1_score(y_test_multi, y_pred_multi, average='macro'):.4f}")

print("\nDetailed Classification Report:")
print(classification_report(y_test_multi, y_pred_multi, 
                           target_names=le_attack.classes_, digits=4))

# Confusion matrix
cm = confusion_matrix(y_test_multi, y_pred_multi)
print("\nConfusion Matrix:")
print(cm)

# ==================== 4. ENSEMBLE APPROACH ====================
print("\n" + "="*60)
print("ENSEMBLE APPROACH")
print("="*60)

# Create ensemble of different models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Prepare features for traditional ML
X_train_ml = X_train_multi_scaled
X_test_ml = X_test_multi_scaled

# Train multiple classifiers
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
svm = SVC(probability=True, random_state=42, class_weight='balanced')
mlp = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=100, random_state=42)

classifiers = [rf, gb, svm, mlp]
names = ['Random Forest', 'Gradient Boosting', 'SVM', 'MLP']

print("\nIndividual Model Performance:")
results = {}
for name, clf in zip(names, classifiers):
    clf.fit(X_train_ml, y_train_multi)
    y_pred = clf.predict(X_test_ml)
    acc = accuracy_score(y_test_multi, y_pred)
    f1 = f1_score(y_test_multi, y_pred, average='macro')
    results[name] = {'accuracy': acc, 'f1_score': f1}
    print(f"{name}: Accuracy={acc:.4f}, F1-Score={f1:.4f}")

# Ensemble predictions (voting)
ensemble_preds = []
for clf in classifiers:
    pred = clf.predict(X_test_ml)
    ensemble_preds.append(pred)

# Simple majority voting
ensemble_vote = np.round(np.mean(ensemble_preds, axis=0)).astype(int)
print(f"\nEnsemble Voting Accuracy: {accuracy_score(y_test_multi, ensemble_vote):.4f}")
print(f"Ensemble Voting F1-Score: {f1_score(y_test_multi, ensemble_vote, average='macro'):.4f}")

# ==================== 5. SAVE MODELS ====================
print("\n" + "="*60)
print("SAVING MODELS")
print("="*60)

# Save models
binary_model.save("binary_model.keras")
multi_class_model.save("multi_class_model.keras")
joblib.dump(scaler_bin, "scaler_bin.pkl")
joblib.dump(scaler_multi, "scaler_multi.pkl")
joblib.dump(le_attack, "label_encoder.pkl")

print("Saved files:")
print("1. binary_model.keras - Binary classification model")
print("2. multi_class_model.keras - Multi-class classification model")
print("3. scaler_bin.pkl - Binary scaler")
print("4. scaler_multi.pkl - Multi-class scaler")
print("5. label_encoder.pkl - Label encoder for attack types")

# ==================== 6. VISUALIZATION ====================
import matplotlib.pyplot as plt
import seaborn as sns

# Plot training history
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Binary model
axes[0, 0].plot(history_bin.history['loss'], label='Train Loss')
axes[0, 0].plot(history_bin.history['val_loss'], label='Val Loss')
axes[0, 0].set_title('Binary Model Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend()

axes[0, 1].plot(history_bin.history['accuracy'], label='Train Acc')
axes[0, 1].plot(history_bin.history['val_accuracy'], label='Val Acc')
axes[0, 1].set_title('Binary Model Accuracy')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Accuracy')
axes[0, 1].legend()

# Multi-class model
axes[1, 0].plot(history_multi.history['loss'], label='Train Loss')
axes[1, 0].plot(history_multi.history['val_loss'], label='Val Loss')
axes[1, 0].set_title('Multi-Class Model Loss')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Loss')
axes[1, 0].legend()

axes[1, 1].plot(history_multi.history['accuracy'], label='Train Acc')
axes[1, 1].plot(history_multi.history['val_accuracy'], label='Val Acc')
axes[1, 1].set_title('Multi-Class Model Accuracy')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Accuracy')
axes[1, 1].legend()

plt.tight_layout()
plt.show()

# Confusion matrix heatmap
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(10, 8))
sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=le_attack.classes_, 
            yticklabels=le_attack.classes_,
            linewidths=0.5)
plt.title('Normalized Confusion Matrix', fontsize=16)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Bar plot of ensemble results
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Accuracy comparison
acc_values = [results[name]['accuracy'] for name in names]
acc_values.append(accuracy_score(y_test_multi, ensemble_vote))
acc_values.append(accuracy_score(y_test_multi, y_pred_multi))

labels = names + ['Ensemble', 'Neural Network']
x_pos = np.arange(len(labels))

axes[0].bar(x_pos, acc_values, alpha=0.8)
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Model Accuracy Comparison')
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels(labels, rotation=45, ha='right')
axes[0].axhline(y=0.8, color='r', linestyle='--', alpha=0.5)

# F1-score comparison
f1_values = [results[name]['f1_score'] for name in names]
f1_values.append(f1_score(y_test_multi, ensemble_vote, average='macro'))
f1_values.append(f1_score(y_test_multi, y_pred_multi, average='macro'))

axes[1].bar(x_pos, f1_values, alpha=0.8, color='orange')
axes[1].set_ylabel('Macro F1-Score')
axes[1].set_title('Model F1-Score Comparison')
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(labels, rotation=45, ha='right')
axes[1].axhline(y=0.8, color='r', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print("1. Binary classification: Excellent (100% accuracy)")
print("2. Multi-class classification (after merging):")
print(f"   - Neural Network Accuracy: {accuracy_score(y_test_multi, y_pred_multi):.4f}")
print(f"   - Neural Network Macro F1: {f1_score(y_test_multi, y_pred_multi, average='macro'):.4f}")
print(f"   - Ensemble Accuracy: {accuracy_score(y_test_multi, ensemble_vote):.4f}")
print(f"   - Ensemble Macro F1: {f1_score(y_test_multi, ensemble_vote, average='macro'):.4f}")
print("3. Attack types after merging:")
for i, class_name in enumerate(le_attack.classes_):
    print(f"   {i}. {class_name}")
print("="*60)
```
